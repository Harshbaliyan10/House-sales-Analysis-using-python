```python
Q1 = data['SalePrice'].quantile(0.25)
Q3 = data['SalePrice'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

no_outliers = data[(data['SalePrice']>= lower_bound) & (data['SalePrice'] <= upper_bound)]
```
